```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic.columns = titanic.columns.str.lower()
```

## Fare (Ticket Price) Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='green')
plt.title('Fare (Ticket Price) Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

## Class vs Survival Rate Plot

```python
plt.figure(figsize=(10,6))
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

## Age Distribution by Class

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=titanic, x='pclass', y='age', palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

## Scatterplot: Age vs Fare

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', palette='coolwarm')
plt.title('Age vs Fare (colored by Survival Status)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

## Countplot: Embarked by Class
* Embarked journey from S - southampton, C - Cherbourg, Q - Queenstown 

```python
plt.figure(figsize=(10,6))
sns.countplot(data=titanic, x='embarked', hue='pclass')
plt.title('Distribution of Passengers by Embarked Port and Class')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```

## Countplot: Survival by Gender

```python
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
